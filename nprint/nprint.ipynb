```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
df_features = pd.read_csv('./deauth.csv')
```

```python
with open('./attack_frames.txt', 'r', encoding='utf-16') as f:
    attack_frames = set(int(line.strip()) for line in f if line.strip())
```

```python
df_features['frame_number'] = range(1, len(df_features) + 1)

df_features['label'] = df_features['frame_number'].apply(lambda x: 1 if x in attack_frames else 0)

print(f"Całkowita liczba ramek: {len(df_features)}")
print(f"Liczba ramek oznaczonych jako atak: {df_features['label'].sum()}")

X = df_features.drop(columns=['label', 'frame_number', 'src_mac', 'dst_mac'], errors='ignore')
y = df_features['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

clf = RandomForestClassifier(
    n_estimators=100, 
    random_state=42,
    n_jobs=-1
)
clf.fit(X_train, y_train)
```

```python
predictions = clf.predict(X_test)
print(classification_report(y_test, predictions))
```
